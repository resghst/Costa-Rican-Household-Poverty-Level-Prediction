```python
# tf_train.rez_esc = tf_train.rez_esc.fillna(0)

data = tf_train.select_dtypes(include=[np.number]).interpolate().fillna(0)
testdata = tf_test.select_dtypes(include=[np.number]).interpolate().fillna(0)
y = tf_train.Target
X = data.drop(['Target'], axis=1)
# eda(tf_test)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=55, test_size=.2)

from sklearn.cluster import KMeans
kmeans = KMeans().fit(X_train,y_train)
predictions = kmeans.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
print ('RMSE is: \t', mean_squared_error(y_test, predictions))
print(tf_test.columns)
feats = tf_test.select_dtypes(include=[np.number]).interpolate().fillna(0)
print(feats.columns)
# eda(feats)
# print(feats.v2a1)
finialpredictions = kmeans.predict(feats)
```

```python
print(tf_test)
print(len(finialpredictions))
submission = pd.DataFrame()
submission['Id'] = tf_test.Id
submission['SalePrice'] = finialpredictions
submission.to_csv('output.csv', index=False)
```
